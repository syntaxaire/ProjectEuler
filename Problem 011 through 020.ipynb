### Problem 12: Highly divisible triangular number
The sequence of triangle numbers is generated by adding the natural numbers. So the 7<sup>th</sup> triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:
```
1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
```
Let us list the factors of the first seven triangle numbers:
```
 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
```
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

````python
# added to euler.py
def factor_list(number):
    """Return a sorted list containing all factors of a positive integer, including 1 and itself."""
    factors = set()
    for i in range(1, int(number ** 0.5) + 1):
        if not(number % i):
            factors.add(i)
            factors.add(number // i)
    factors = list(factors)
    factors.sort()
    return factors
````

````python
# added to euler.py
def gen_triangular_numbers():
    tri = 0
    seq = 0
    while True:
        seq = seq + 1
        tri = tri + seq
        yield tri
````

````python
mostdivisors = 0
for i in gen_triangular_numbers():
    factors = factor_list(i)
    if len(factors) > mostdivisors:
        print("Found new triangular number with most divisors:")
        print(" {} with {} divisors: {}".format(i, len(factors), factors))
        mostdivisors = len(factors)
    if len(factors) > 500:
        break
````

### Problem 14: Longest Collatz sequence
The following iterative sequence is defined for the set of positive integers:
$$
n → n/2\ (n\ is\ even)\\
n → 3n + 1\ (n\ is\ odd)
$$
Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

````python
# added to euler.py
def gen_collatz(n):
    """Yield the Collatz sequence beginning with n, one term at a time."""
    yield n
    while n != 1:
        if n % 2:
            n = n * 3 + 1
            yield n
        else:
            n //= 2
            yield n
````

````python
longest = 0
winner = 0
for i in range(1, 1000000):
    newcollatz = list(gen_collatz(i))
    if len(newcollatz) > longest:
        winner = i
        longest = len(newcollatz)
        print("New longest chain found with {}, new length {}.".format(winner, len(newcollatz)))
````

### Problem 15: Lattice paths
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.
![diagram](http://projecteuler.net/project/images/p015.gif)
How many such routes are there through a 20×20 grid?

#### Work
For every column in the grid, there are height+1 paths through it (the +1 is the line which underscores the column). For a path taken through any column, the number of paths available in the next column is equal to the height+1 of the path we just took.

````python
"""Return the number of paths down and to the right through a x by y grid.
Exponentially slower with large sizes."""
def gridpaths(xsize, ysize):
    if xsize == 1: # last column
        return ysize + 1 # height + 1
    total = 0
    for i in range(0, ysize + 1):
        total += gridpaths(xsize - 1, i)
    return total
````

````python
for i in range(1, 11):
    print("The number of paths through a {}x{} grid is {}.".format(i, i, gridpaths(i, i)))
````

This function becomes exponentially slower noticeable around size (15, 15) so probably unsuitable for (20, 20).

However, by running this function to 10 we have discovered that the sequence is the [central binomial coefficients](https://oeis.org/A000984).

````python
"""Return the list of terms to the nth term from the sequence of central binomial coefficients."""
def centralbinomialcoefficient(n):
    cbclist = [1]
    term = 1
    for i in range(n):
        term = term * (4 * i + 2) // (i + 1)
        cbclist.append(term)
    print(cbclist)
````

### Problem 17: Number letter counts
If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are $3 + 3 + 5 + 4 + 4 = 19$ letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

````python
names = {1:'one',2:'two',3:'three',4:'four',5:'five',6:'six',7:'seven',8:'eight',9:'nine',
         10:'ten',11:'eleven',12:'twelve',13:'thirteen',14:'fourteen',15:'fifteen',
         16:'sixteen',17:'seventeen',18:'eighteen',19:'nineteen',
         20:'twenty',30:'thirty',40:'forty',50:'fifty',60:'sixty',70:'seventy',80:'eighty',90:'ninety',
         100:'hundred',1000:'thousand',1000000:'million',1000000000:'billion',1000000000000:'trillion'}
def pronounce(number):
    result = ""
    if number > 999:
        result += names[int(str(number)[0])] + ' thousand'
        number -= int(str(number)[0]) * 1000
    if number > 99:
        result += names[int(str(number)[0])] + ' hundred'
        number -= int(str(number)[0]) * 100
    if number > 0 and len(result) > 0:
        result += ' and '
    if number in names:
        result += names[number]
        number -= number # skips the next if, if the <100 number had a specific name
    if number > 0: # number is between 1 and 99 inclusive and is a compound name (21-99 but not a multiple of ten)
        result += names[int(str(number)[0]) * 10]
        number -= int(str(number)[0]) * 10
    if number != 0:
        result += ' ' + names[number]
    return result
````

````python
bigresult = ""
for i in range(1, 1001):
    for j in pronounce(i).split(' '):
        bigresult += j
print(len(bigresult))
````

Find the maximum total from top to bottom of the triangle below:

````python
tristring ="""75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23"""
# build a list that looks just like what we see above
tri = [[int(char) for char in line.split(' ')] for line in tristring.split('\n')]
def maxsum(line, col):
    if line == len(tri) - 1: # are we in the bottom line?
        return tri[line][col] # yes, just return the number here
    # otherwise, return the number here plus the largest recursive option from below
    return tri[line][col] + max(maxsum(line + 1, col), maxsum(line + 1, col + 1))
print(maxsum(0, 0))
````

### Problem 19: Counting Sundays

You are given the following information, but you may prefer to do some research for yourself.

- 1 Jan 1900 was a Monday.
- Thirty days has September,
- April, June and November.
- All the rest have thirty-one,
- Saving February alone,
- Which has twenty-eight, rain or shine.
- And on leap years, twenty-nine.
- A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?


````python
# Using the standard library
import calendar
numsundays = 0
for year in range(1901, 2000 + 1):
    for month in range(1, 12 + 1):
        if calendar.weekday(year, month, 1) == 6: # 0 is Monday, 6 is Sunday (European convention)
            numsundays += 1
print(numsundays)
````
